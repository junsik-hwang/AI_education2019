### 9. score.txt 파일의 내용을 불러온 후 다음을 계산하시오.
- 자료는 앞에서부터 학번, 영어, 국어, 수학, 체육, 컴퓨터의 시험점수이다.
- 전체 평균점수가 가장 높은 학생과 가장 낮은 학생은 각각 몇 번 학생인가?
- 국어, 영어, 수학 점수의 평균이 가장 높은 학생은 누구인가?
- 국어, 영어, 수학 점수에 2배 가중치를 준다면 가장 점수가 높은 학생은 누구인가?

```python
score = np.loadtxt('score.txt')
score
```

```python
score1 = np.zeros((9,2))
score2 = np.zeros((9,2))
```

```python
max_student_score = 0
min_student_score = 100
for i in range(1,10):
    score1[i-1][0] = i
    for j in range(1,6):
        score1[i-1][1] = score1[i-1][1] + score[i-1][j]
    score1[i-1][1] = score1[i-1][1] / 5
    if(max_student_score < score1[i-1][1]):
        max_student_score = score1[i-1][1]
        max_student = score1[i-1][0]
    if(min_student_score > score1[i-1][1]):
        min_student_score = score1[i-1][1]
        min_student = score1[i-1][0]
```

```python
print("평균점수가 가장 높은 학생 : {0}번 학생(점수 : {1})".format(max_student,max_student_score))
print("평균점수가 가장 낮은 학생 : {0}번 학생(점수 : {1})".format(min_student,min_student_score))
```

```python
max_student_score = 0
for i in range(1,10):
    score2[i-1][0] = i
    for j in range(1,4):
        score2[i-1][1] = score2[i-1][1] + score[i-1][j]
    score2[i-1][1] = score2[i-1][1] / 3
    if(max_student_score < score2[i-1][1]):
        max_student_score = score2[i-1][1]
        max_student = score2[i-1][0]

print("국 영 수 평균점수가 가장 높은 학생 : {0}번 학생(점수 : {1})".format(max_student,max_student_score))
```
